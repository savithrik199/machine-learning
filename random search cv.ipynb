```python
import random
dist={"max_depth":[3,5,10,None],
     "n_estimators":[10,20,30,40,50],
     "max_features":["auto","sqrt"],
     "criterion":["gini","entropy"],
     "bootstrap":[True,False],
     "min_samples_leaf":[1,2]
     "min_samples_split":[2,5]}
```
